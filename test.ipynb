```python
import face_recognition as fr
import api
from skimage import io as io
from matplotlib import pyplot as plt
import timeit

img1_r = io.imread('test1.jpg')
t1 = timeit.default_timer()
img1_a = api.face_alignment(img1_r, scale = 1.05)
print('Time elapsed: {}'.format(timeit.default_timer() - t1))
print('Original image')
plt.imshow(img1_r)
plt.show()
print('After alignment')
plt.imshow(img1_a[0])
plt.show()

img2_r = io.imread('test2.jpg')
t1 = timeit.default_timer()
img2_a = api.face_alignment(img2_r, scale = 1.05)
print('Time elapsed: {}'.format(timeit.default_timer() - t1))
print('Original image')
plt.imshow(img2_r)
plt.show()
print('After alignment')
plt.imshow(img2_a[0])
plt.show()

img3_r = io.imread('test3.jpg')
t1 = timeit.default_timer()
img3_a = api.face_alignment(img3_r, scale = 1.05)
print('Time elapsed: {}'.format(timeit.default_timer() - t1))
print('Original image')
plt.imshow(img3_r)
plt.show()
print('After alignment')
plt.imshow(img3_a[0])
plt.show()


img4_r = io.imread('test4.jpg')
t1 = timeit.default_timer()
img4_a = api.face_alignment(img4_r, scale = 1.05)
print('Time elapsed: {}'.format(timeit.default_timer() - t1))
print('Original image')
plt.imshow(img4_r)
plt.show()
print('After alignment')
for img in img4_a:
    plt.imshow(img)
    plt.show()
```
